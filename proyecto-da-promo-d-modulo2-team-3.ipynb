```python
print(artists)
```

```python
artista_fm = []

for artist in artists:
    params_info = {
        'method': 'artist.getinfo',
        'artist': artist,  
        'api_key': API_KEY,
        'format': 'json'
    }
    
    response_info = requests.get(BASE_URL, params=params_info)
    data_info = response_info.json()
    
    if response_info.status_code != 200:
        print(f"Error en la solicitud para {artist}: {response_info.status_code}")
        continue

    else:
        artist_data = data_info.get('artist',{})
        summary = artist_data.get('bio',{}).get('summary', 'Información no disponible')
        listeners = artist_data.get('stats', {}).get('listeners', 'No disponible')
        playcount = artist_data.get('stats', {}).get('playcount', 'No disponible')
        similar_artist = [sim['name'] for sim in artist_data.get('similar', {}).get('artist', [])]
        artista_fm.append({'Artist': artist,
                            'Bio': summary,
                            'Listeners': listeners,
                            'Playcount': playcount,
                            'Similar Artists': similar_artist})
```

```python
df_artista_fm = pd.DataFrame(artista_fm)
```
